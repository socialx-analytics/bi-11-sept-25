```python
# 1. Import library
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
# 2. Ambil konten website
url = "https://www.cnbcindonesia.com/search?query=syariah"
response = requests.get(url)
response
```

```python
# 3. Parse HTML
soup = BeautifulSoup(response.text, 'html.parser')
soup
```

```python
# 4. Gunakan selector yang didapat dari plugin
berita_list = soup.select('.group-hover\:text-cnbc-primary-blue')
berita_list
```

```python
# 5. Menampilkan hasil
judul_list = []
for berita in berita_list:
    judul = berita.get_text().strip()
    print(judul)
    judul_list.append(judul)
```

```python
# 6. Dijadikan pandas DataFrame & tampilkan DataFramenya
df = pd.DataFrame({'Judul Berita': judul_list})
print(df)
```

```python
# 7. Simpan ke CSV
df.to_csv('daftar_berita_syariah.csv', index=False)
```
