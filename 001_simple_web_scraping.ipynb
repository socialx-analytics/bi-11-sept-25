```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
from urllib.parse import quote_plus
from google.colab import files
```

```python
# Set query & URL
query = "ekonomi indonesia"   # ubah sesuai kebutuhan
url = f"https://www.cnbcindonesia.com/search?query={quote_plus(query)}"
url
```

```python
# Ambil konten website
headers = {"User-Agent": "Mozilla/5.0"}
response = requests.get(url, headers=headers)
response
```

```python
# Parse HTML
soup = BeautifulSoup(response.text, "lxml")
soup
```

```python
# Ambil elemen judul berita
berita_list = soup.select(".group-hover\\:text-cnbc-primary-blue")
berita_list
```

```python
# Simpan judul ke list
judul_list = [b.get_text(strip=True) for b in berita_list]
for j in judul_list:
    print(j)
```

```python
# Buat DataFrame
df = pd.DataFrame({"Judul Berita": judul_list})
df
```

```python
# Simpan CSV (nama file otomatis sesuai query) & auto-download
filename = f"berita_cnbc_{query.replace(' ', '_').lower()}.csv"
df.to_csv(filename, index=False, encoding="utf-8-sig")
files.download(filename)
```
