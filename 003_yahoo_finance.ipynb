```python
# Install & import library
# !pip -q install yfinance

import yfinance as yf
import pandas as pd
from google.colab import files
```

```python
# Set simbol saham/indeks
symbol = "^JKLQ45"   # LQ45 Index
symbol
```

```python
# Ambil data harga (default: periode max, interval 1 hari)
df = yf.download(symbol, period="max", interval="1d")
df
```

```python
# Simpan ke CSV & auto-download
filename = f"{symbol.replace('^','')}.csv"
df.to_csv(filename)
files.download(filename)
```
