```python
import yfinance as yf
import pandas as pd
```

```python
def get_price_data(symbol, period='max', interval='1d'):
    data = yf.download(symbol, period=period, interval=interval)
    return data
```

```python
symbol = "^JKLQ45"
```

```python
df = get_price_data(symbol)
df.to_csv(f"{symbol}.csv")
```
